```python
import math, nltk

nltk.download('brown')
nltk.download('punkt')


number_of_texts_to_keep = 20 ### SET THE NUMBER OF FILES YOU WANT TO KEEP ###

from nltk.corpus import brown 

files = nltk.corpus.brown.fileids()
files = files[:number_of_texts_to_keep]
print(files)
all_texts_words = brown.words(files)
print(all_texts_words)
```

```python
print('We have', len(files), 'text files,')
print('which contain', len(brown.words(files)), 'words')
```

```python
from nltk import sent_tokenize
from nltk import word_tokenize

all_texts_string = " ".join(all_texts_words)
sentences = sent_tokenize(all_texts_string)

for sent in sentences:
    sent_tok = word_tokenize(sent)
            
    print(sent_tok)
    print("_________________")
```

```python
from pprint import pprint

N = 20
# Frequency distribution
count = nltk.FreqDist(all_texts_words)

unks = 0
count_unk = count.copy()
garbage_toks = ['.', ',', '--', '(', ')', '$', '``', '`', '-', '?', ':', "'", ';', "''"]

for item in count.items():
    if item[0] in garbage_toks:
        del count_unk[item[0]]
    elif item[1] < 10:
        count_unk.update({'*UNK*': item[1]})
        del count_unk[item[0]]      
        unks += 1

vocab_size = len(count_unk.items())

print(f'Found {unks} unknown words\n')

print(f'We had {count.N()} tokens with a vocabulary of size {len(count.items())}\n')

print(f'After setting rare tokens to *UNK* we have a vocabulary of size {vocab_size}\n')

print('Most common tokens: \n')
pprint(count_unk.most_common(N))
```

```python
from collections import Counter
from nltk.util import ngrams
from pprint import pprint

sentences_tokenized = []
vocab = list(count_unk.keys())

for sent in sentences:
    sent_tok = word_tokenize(sent)
    ## find *UNK* and garbage chars 
    temp_toks = list(map(lambda tok: tok if tok in vocab and not (tok in garbage_toks) else ('*UNK*' if not tok in garbage_toks else None), sent_tok))
    res_toks = list(filter(lambda item: item is not None, temp_toks))
    sentences_tokenized.append(res_toks)
    
    print(res_toks)
    print("_________________")




unigram_counter = Counter()
bigram_counter = Counter()
trigram_counter = Counter()

for sent in sentences_tokenized:
    
    unigram_counter.update([gram for gram in ngrams(sent, 1, pad_left=True, pad_right=True,
                                                   left_pad_symbol='*start*',right_pad_symbol='*end') ])
    bigram_counter.update([gram for gram in ngrams(sent, 2, pad_left=True, pad_right=True,
                                                   left_pad_symbol='*start*',right_pad_symbol='*end*') ])
    trigram_counter.update([gram for gram in ngrams(sent, 3, pad_left=True, pad_right=True,
                                                   left_pad_symbol='*start*',right_pad_symbol='*end*') ])


pprint(unigram_counter.most_common(10))   
pprint(bigram_counter.most_common(10))
pprint(trigram_counter.most_common(10))
```

```python
import math

alpha = 0.1

bigram_probs = dict()
bigram_log_probs = dict()

trigram_probs = dict()
trigram_log_probs = dict()

for bigram in bigram_counter:
    bigram_probs[bigram] = (bigram_counter[bigram] + alpha) / (unigram_counter[(bigram[0],)] + alpha*vocab_size)
    bigram_log_probs[bigram] = math.log2(bigram_probs[bigram])
    # print(bigram)
    # print("bigram_prob: {0:.5f} of bigram ".format(bigram_probs[bigram]))

for trigram in trigram_counter:
    trigram_probs[trigram] = (trigram_counter[trigram] + alpha) / (bigram_counter[(trigram[0], trigram[1])] + alpha*vocab_size)
    trigram_log_probs[trigram] = math.log2(trigram_probs[trigram])
    # print(trigram)
    # print("trigram_prob: {0:.5f} of trigram ".format(trigram_probs[trigram]))
```

## Cross entropies and perplexities

### Bigram LM cross entropy and perplexity

```python
sum_prob = 0
bigram_cnt = 0

for sent in sentences_tokenized:
    sent = ['*start*'] + sent + ['*end*']
    bigram_cnt += 1 # end tokens
    for idx in range(1, len(sent)-1):
        sum_prob += math.log2(bigram_probs[(sent[idx], sent[idx+1])])
        bigram_cnt += 1

HC = -sum_prob / bigram_cnt
perpl = math.pow(2,HC)
print("Cross Entropy: {0:.3f}".format(HC))
print("perplexity: {0:.3f}".format(perpl))
```

### Trigram LM cross entropy and perplexity

```python
sum_prob = 0
trigram_cnt = 0

for sent in sentences_tokenized:
    sent = ['*start*'] + ['*start*'] + sent + ['*end*'] + ['*end*']
    trigram_cnt += 2 # end tokens
    for idx in range(2, len(sent)-2):
        sum_prob += math.log2(trigram_probs[(sent[idx], sent[idx+1], sent[idx+2])])
        trigram_cnt += 1

HC = -sum_prob / trigram_cnt
perpl = math.pow(2,HC)
print("Cross Entropy: {0:.3f}".format(HC))
print("perplexity: {0:.3f}".format(perpl))
```

```python
def beam_search_decoder(text, l1):
    k = len(text)
    l2 = 1 - l1
    LK = []
    
    def Lk(l1, l2, t1, t2, w1, w2):
      if (t1, t2) in bigram_log_probs.keys():
          Pt1 = bigram_log_probs[(t1, t2)]
      else:
          print("")
          bigram_probs[(t1, t2)] = alpha / (unigram_counter[(t1,)] + alpha*vocab_size)
          bigram_log_probs[(t1, t2)] = math.log2(bigram_probs[(t1, t2)])
          Pt1 = bigram_log_probs[(t1, t2)]

      # the 1 in the denominators is to cope with identic tokens
      Pw1t1 = math.log2(1/(nltk.edit_distance(t1, w1) + 1)) + math.log2(1/(nltk.edit_distance(t2, w2) + 1))
      return -l1 * Pt1 -l2 * Pw1t1
    

    paths = [[0, '*start*'], [0, '*start*']]
    prev_word = '*start*'
    for word in text[1:]:
      sorted_by_dist = vocab.copy()
      sorted_by_dist.sort(key=lambda w: nltk.edit_distance(word, w))
      sorted_by_dist = sorted_by_dist[:10]
      best1 = (paths[0][0] + Lk(l1, l2, paths[0][1], sorted_by_dist[0], word, prev_word), sorted_by_dist[0])
      best2 = (paths[1][0] + Lk(l1, l2, paths[1][1], sorted_by_dist[1], word, prev_word), sorted_by_dist[1])
      best = (best1, best2)
      from0 = 0 # this will track on which path we are
      from1 = 1
      i = -1    # number of path
      for path in paths:
        i += 1
        for candidate in sorted_by_dist:
          cand_score = path[0] + Lk(l1, l2, path[1], candidate, word, prev_word) 
          if best[0][1] != candidate: # this is in order to really keep the two best and not two times the best one
              if cand_score < best[0][0]:
                  temp = best[0]
                  best = ((cand_score, candidate), best[1])
                  from0 = i
                  if temp[0] < best[1][0]:
                    best = (best[0], temp)
                    from1 = i
              elif cand_score < best[1][0]:
                  best = (best[0], (cand_score, candidate))
                  from1 = i
      prev_word = word

      if from0 == 0 and from1 == 1:
        paths[0][0] = best[0][0]
        paths[0].insert(1, best[0][1])
        paths[1][0] = best[1][0]
        paths[1].insert(1, best[1][1])
      elif from0 == 0 and from1 == 0:
        paths[0][0] = best[0][0]
        paths[0].insert(1, best[0][1])
        paths[1] = paths[0].copy()
        paths[1][0] = best[1][0]
        paths[1][1] = best[1][1]
      elif from0 == 1 and from1 == 1:
        paths[0] = paths[1].copy()
        paths[0][0] = best[0][0]
        paths[0].insert(1, best[0][1])
        paths[1][0] = best[1][0]
        paths[1].insert(1, best[1][1])
      elif from0 == 1 and from1 == 0:
        paths[1][0] = best[1][0]
        paths[1].insert(1, best[1][1])
        paths[0][0] = best[0][0]
        paths[0].insert(1, best[0][1])

    if paths[0][0] < paths[1][0]:
      result = paths[0][1:]
    else: 
      result = paths[1][1:]
    result.reverse()
    return result
      

res = beam_search_decoder(['*start*', 'I', 'am', 'really', 'sorry', 'for', 'you'], 0.5)

print(res)
```
